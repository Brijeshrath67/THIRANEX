```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_wine_data.csv")

df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df["target"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x=df["target"])

plt.show()
```

```python
correlation = df.corr()

correlation
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    correlation,
    cmap="coolwarm"
)

plt.show()
```

```python
top_corr = correlation["target"].sort_values(
    ascending=False
)

top_corr
```

```python
top_features = top_corr.index[1:6]

top_features
```

```python
for col in top_features:
    
    plt.figure(figsize=(6,4))
    
    sns.histplot(
        data=df,
        x=col,
        hue="target",
        kde=True
    )
    
    plt.show()
```

```python
sns.pairplot(
    df,
    hue="target",
    vars=top_features
)

plt.show()
```

```python
mean_values = df.groupby("target").mean()

mean_values
```

```python
mean_values.T.plot(
    figsize=(14,8)
)

plt.show()
```

```python
variance = df.var()

variance.sort_values(
    ascending=False
)
```
